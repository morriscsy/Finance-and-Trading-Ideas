```python
df = pd.DataFrame(historical_data)
del df["date"]
df.rename(columns={'formatted_date':'date'},inplace=True)
df.index = pd.DatetimeIndex(df["date"].values)
```

```python
df.head()
```

```python
df.loc["2009-01-02":"2009-02-07",["high","low"]]
```

use date to index 

```python
df["close_7days_sma"] = df["close"].rolling(window=7).mean()
df["close_20days_sma"] = df["close"].rolling(window=20).mean()
```

```python
df.head(21)
```

```python
df["20std"]=df["close"].rolling(window=20).std()
df["bollinger_upper"] = df["close_20days_sma"] + 2*df["20std"]
df["bollinger_lower"] = df["close_20days_sma"] - 2*df["20std"]
df.dropna(inplace=True)
df.head()
```

```python
df.loc["2017-11-01":"2018-12-31",['close',"close_20days_sma",'bollinger_upper','bollinger_lower']].plot(figsize=(12,7),kind="line",title="Bollinger Bands", legend=True)
```
